```python
import cv2
import matplotlib.pyplot as plt
```

```python
config_file="ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt"
frozen_model="frozen_inference_graph.pb"
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLabels = [] ## empty List of python file_name = 'Labels.txt'
file_name="labels.txt"

with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(classLabels)
```

```python
model.setInputSize(320, 320) 
model.setInputScale(1.0/127.5) ## 255/2 = 127.5 
model.setInputMean((127.5,127.5,127.5)) ## mobilenet => [-1,1] 
model.setInputSwapRB(True)
```

```python
cap1 = cv2.VideoCapture("4p-c0.avi")

# Check if the video is opened correctly

if not cap1.isOpened():
    cap1 = cv2.VideoCapture(0)

if not cap1.isOpened():
    raise IOError("Cannot open video")

pre=0
flag=0
font_scale = 1
font = cv2.FONT_HERSHEY_PLAIN
pathOut="try1.avi"
fps=25
size= (360,288)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
threshold=0.01
num_c0=0
num_c1=0
num_c2=0

while (cap1.isOpened()):

    ret, frame = cap1.read()
    num_c0+=1
    if frame is not None:
        c=[]
        ClassIndex, confidece, bbox = model.detect(frame, confThreshold=0.55)
        print (ClassIndex,confidece)

        if (len(ClassIndex)!=0):
            for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
                if (ClassInd==1):
#                     print(type(frame))
#                     print(frame.shape)
                    c.append(conf)
                    cv2.imwrite("./Frame/_"+ str(num_c0) + "_c0.png", frame)
                    flag=1
                    
#                     cv2.rectangle(frame, boxes, (255, 0, 0), 2)
#                     cv2.putText(frame,classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0 ,255, 0))
                    
            if flag==1:
                
                curr=sum(c)
                if curr-pre>threshold:
                    out.write(frame)
                    
                pre = curr
                    
                    
        cv2.imshow('Object Detection', frame)

        if cv2.waitKey(2) & 0xFF == ord('q'):
            break

    else:
        break

print("END")
out.release()
cap1.release()
cv2.destroyAllWindows()  
```

```python
import cv2
import glob

pre=0
flag=0
font_scale = 1
font = cv2.FONT_HERSHEY_PLAIN
pathOut="output.mp4"
fps=10
size= (480,848)
out = cv2.VideoWriter(pathOut,cv2.VideoWriter_fourcc(*'DIVX'), fps, size)
threshold=0.01

for img in glob.glob("./<>/*.png"):
    cv_img = cv2.imread(img)
    ClassIndex, confidece, bbox = model.detect(cv_img, confThreshold=0.55)
    
    if (len(ClassIndex)!=0):
            for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
                if (ClassInd==1):
                    c.append(conf)
                    flag=1
                    
                    cv2.rectangle(cv_img, boxes, (255, 0, 0), 2)
                    cv2.putText(cv_img,classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0 ,255, 0))
                    
            if flag==1:
                
                curr=sum(c)
                if curr-pre>threshold:
                    out.write(cv_img)
                    
                pre = curr
out.release()
```
